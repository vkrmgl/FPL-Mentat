```python
import duckdb
```

```python
con = duckdb.connect('../data/fpl-mentat.duckdb')
```

```python
print(con.execute("select * from stg_players").fetchdf())
```

```python
df = con.execute("""
    select player_id, second_name, team_id
    from stg_players
    where lower(second_name) like '%semenyo%'
        or lower(second_name) like '%larsen%'
        or lower(second_name) like '%bobb%'
        or lower(second_name) like '%traore%'
""").fetchdf()
```
